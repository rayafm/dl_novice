```python
import tensorflow as tf
```

```python
# Data fashion MNIST didapatkan dengan mudah dari library datasets yang disediakan Keras
mnist = tf.keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Dalam klasifikasi gambar, setiap piksel pada gambar memiliki nilai dari 0 sampai 255. 
# Perlu melakukan normalisasi dengan membagi setiap pixel pada gambar dengan 255. 
# Dengan nilai yang telah dinormalisasi, jaringan saraf dapat belajar dengan lebih baik.
X_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
# Mendefinisikan arsitektur dari jaringan saraf yang akan dilatih 
# 3 layer utama:  input layer, hidden layer, dan output layer
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)), 
    tf.keras.layers.Dense(512, activation=tf.nn.relu), 
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
# Salah satu optimizer yang biasa digunakan adalah Adam  
# Selanjutnya untuk loss function dapat menggunakan sparse categorical entropy pada kasus klasifikasi 3 kelas atau lebih 
# Parameter metrics berfungsi untuk menampilkan metrik yang dipilih pada proses pelatihan model
model.compile(optimizer=tf.optimizers.Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10)
```
